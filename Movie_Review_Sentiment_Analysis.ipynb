```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
data = {
    'Review': [
        'This movie was amazing',
        'Worst movie ever',
        'I really liked this film',
        'Very boring and slow',
        'Excellent acting and story',
        'Terrible experience',
        'Fantastic movie',
        'Not good',
        'Loved the cinematography',
        'Waste of time'
    ]
}

df = pd.DataFrame(data)

df
```

```python
def get_sentiment(text):

    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        return 'Positive'

    elif analysis.sentiment.polarity < 0:
        return 'Negative'

    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Review'].apply(get_sentiment)

df
```

```python
df['Sentiment'].value_counts().plot(kind='bar')

plt.title("Movie Review Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

# Insights

1. Positive movie reviews are higher than negative reviews.

2. Sentiment Analysis helps in understanding audience opinions.

3. Python libraries can classify text data based on sentiment polarity.

4. Data visualization makes sentiment distribution easier to understand.

**Dataset for Sentiment Analysis**

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IMDB Dataset.csv")

df.head()
```

```python
df = df.head(100)
```

```python
def get_sentiment(text):

    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        return 'Positive'

    elif analysis.sentiment.polarity < 0:
        return 'Negative'

    else:
        return 'Neutral'
```

```python
df['Predicted_Sentiment'] = df['review'].apply(get_sentiment)

df[['review', 'sentiment', 'Predicted_Sentiment']].head()
```

```python
df['Predicted_Sentiment'].value_counts().plot(kind='bar')

plt.title("Sentiment Analysis on IMDb Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

# Insights

1. Most movie reviews were classified as positive.

2. Sentiment analysis helps in understanding audience reactions.

3. TextBlob can classify textual data based on sentiment polarity.

4. Visualization helps in understanding sentiment distribution clearly.
